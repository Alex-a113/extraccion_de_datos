```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/Datos_Facturaci√≥n.csv')
df = df.drop('Unnamed: 0', axis=1)
display(df.head())
```

```python
f1 = df[(df['CVE_CLPV'] >= 1000) & (df['CVE_CLPV'] <= 2000)]
display(f1)
'''
Estamos utilizando la variable CVE_CLPV en la cual la filtramos que sea es mayor o igual a 1000 o es menor o igual a 2000.
'''
```

```python
f1.to_csv('Filtro1-CVE_PLV.csv', index=False)
```

```python
f2 = df_filtrado = df[~((df["CVE_VEND"] == 4) | (df["CVE_VEND"] == 5))]
display(f2)

'''
Utilizamos la variable CVE_VEND para que no muestre los valores 4 y 5.
'''
```

```python
f2.to_csv('Filtro2-CVE_VEND.csv', index=False)
```

```python
df["FECHA_ENT"] = pd.to_datetime(df["FECHA_ENT"], dayfirst=True, errors='coerce')
filtro_fecha = df[df["FECHA_ENT"] == pd.to_datetime("2022-02-28")]
display(filtro_fecha)

'''
Filtramos la variable FECHA_ENT para que muestre solo el dia 28 de febrero de
2022 y primero la convertimos a datetime para evitar errores
'''
```

```python
filtro_fecha.to_csv('Filtro3-FECHA_ENT.csv', index=False)
```

```python
f4 = df[(df["CAN_TOT"] < 5951.7) | (df["STATUS"] == "E")]
display(f4)

'''
Filtramos las dos variables CAN_TOT y STATUS en la cual filtramos las filas
< 5951.7) o donde el estado es ‚ÄúE‚Äù
'''
```

```python
f4.to_csv('Filtro4-CAN_TOT.csv', index=False)
```

```python
f5 = df[["CVE_DOC", "FECHA_ENT", "FECHA_VEN", "CAN_TOT"]]
display(f5.head(8))

'''
Filtramos las variables CVE_DOC, FECHA_ENT, FECHA_VEN, CAN_TOT en un nuevo dataframe
'''
```

```python
f5.to_csv('Filtro5-COLUMNAS.csv', index=False)
```

```python
f6 = df.iloc[7001:7100]
display(f6.head(8))

'''
Filtramos las filas desde la 7001 a la 7100
'''
```

```python
f6.to_csv('Filtro6-FILAS.csv', index=False)
```

```python
df = df[df["CVE_VEND"].isin([1, 2])]
f7 = df[["CVE_VEND", "FECHAELAB"]].set_index("CVE_VEND")

print(f7)

'''
FIltramos la variable CVE_VEND en la cual muestre solo los valores 1 y 2.
Seleccionamos las columnas CVE_VEND y FECHAELAB y las mostramos en un nuevo dataframe con el indice CVE_VEND
'''
```
